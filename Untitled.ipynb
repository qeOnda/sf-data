```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('https://data.sfgov.org/resource/5cei-gny5.csv')
fd = pd.read_csv('https://data.sfgov.org/resource/wmam-7g8d.csv')
```

```python
eviction_zip = df[['eviction_id', 'zip']].groupby(by = 'zip', as_index=False).count().sort_values('eviction_id', ascending=False)
eviction_zip.rename(columns = {'eviction_id': 'evictions'}, inplace=True)
eviction_zip.head(6)
```

```python
eviction_hood = df[['eviction_id', 'neighborhood']].groupby(by = 'neighborhood', as_index=False).count().sort_values('eviction_id', ascending=False)
eviction_hood.rename(columns = {'eviction_id': 'evictions', 'neighborhood': 'neighborhood'}, inplace=True)
eviction_hood.head(10)
```

```python
alt.Chart(eviction_hood).mark_bar().encode(x = alt.X('neighborhood', sort='-y'), y= 'evictions') 
```

```python
evict_months = df.eviction_id.groupby(by = df.file_date.dt.month).count()
evict_months
```

```python
evict_day = df[['eviction_id', 'file_date']].groupby(by = 'file_date', as_index=False).count()
evict_day.rename(columns = {'eviction_id': 'evictions'}, inplace=True)
evict_day.head()
```

```python
alt.Chart(
    evict_day,
    padding = 10,
    title = 'Evictions by day of year'
).mark_line(
    interpolate = 'basis',
    color = 'blue'
).encode(
    x = alt.X(
        'file_date:T',
        axis = alt.Axis(
            title = 'Day of year',
            format = '%b-%e',
            labelAngle = -90, 
            labelFlush = True, 
            tickCount = 26, 
            grid = False)),
        y = alt.Y(
            'evictions:Q',
            axis =alt.Axis(
            title = 'Evictions'))
    )
        
```
